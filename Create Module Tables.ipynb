```python
import pandas as pd 
```

```python
# load module

data = pd.read_csv('TCGA Barcodes/Modules/LUAD_M1.csv')

data
```

```python
# load NFE genes, for LUAD

LUAD_NFE = pd.read_csv('LUAD NFE + Median.csv', sep=",")

LUAD_NFE
```

```python
# load TracerX genes

TracerX = pd.read_csv('TCGA Barcodes\Modules\TracerXGenes.csv')

TracerX
```

```python
# load CGC genes

CGC = pd.read_csv('TCGA Barcodes\Modules\CGC Cancer Genes Tier 1 + 2 - Clean.csv')

CGC
```

```python
# load KEGG metabolism genes 

KEGG_M = pd.read_csv('TCGA Barcodes\Modules\KEGG Pathways - Metabolism.csv')

KEGG_M
```

```python
# test count with NFE

NFE_count = (LUAD_NFE['HUGO_SYMBOL']).isin(data['LUAD_M1'])

NFE_count.sum()
```

```python
# test count with TracerX

TracerX_count = TracerX['Tracer X Genes'].isin(data['LUAD_M1'])

TracerX_count.sum()
```

```python
# summary of LUAD modules 
# note: could clean up and put with LUSC.

LUAD_module = ['LUAD_M1', 'LUAD_M2', 'LUAD_M3', 'LUAD_M4']

for item in LUAD_module: 
    data = pd.read_csv('TCGA Barcodes/Modules/' + item + '.csv')
    NFE_count = LUAD_NFE['HUGO_SYMBOL'].isin(data[item])
    TracerX_count = TracerX['Tracer X Genes'].isin(data[item])
    CGC_count = CGC['CGC Gene'].isin(data[item])
    gene_count = data.size
    KEGG_count = KEGG_M['GENE'].isin(data[item])
    
    print (item, gene_count,'genes,', NFE_count.sum(), 'NFE genes,', TracerX_count.sum(), 'Tracer X genes,', 
           CGC_count.sum(), 'CGC genes,', KEGG_count.sum() , 'KEGG Metabolism genes', '\n')
```

```python
# LUSC

LUSC_module = ['LUSC_M1', 'LUSC_M2', 'LUSC_M3', 'LUSC_M4', 'LUSC_M5', 'LUSC_M6']
LUSC_NFE = pd.read_csv('LUSC NFE + Median.csv', sep=",")


for item in LUSC_module: 
    data = pd.read_csv('TCGA Barcodes/Modules/' + item + '.csv')
    NFE_count = (LUSC_NFE['HUGO_SYMBOL']).isin(data[item])
    TracerX_count = TracerX['Tracer X Genes'].isin(data[item])
    CGC_count = CGC['CGC Gene'].isin(data[item])
    gene_count = data.size
    KEGG_count = KEGG_M['GENE'].isin(data[item])
    
    print (item, gene_count,'genes,', NFE_count.sum(), 'NFE genes,', TracerX_count.sum(), 'Tracer X genes,', 
           CGC_count.sum(), 'CGC genes,', KEGG_count.sum() , 'KEGG Metabolism genes', '\n')
```

```python
# do for all LUAD 

LUAD_module = ['LUAD_M1', 'LUAD_M2', 'LUAD_M3']

for item in LUAD_module: 
    data = pd.read_csv('TCGA Barcodes/Modules/' + item + '.csv')
    NFE_genes = (data[item]).isin(LUAD_NFE['HUGO_SYMBOL'])
    data['NFE Gene'] = NFE_genes
    NFE_True = data.loc[data['NFE Gene'] == True]
    
    Max_dict = LUAD_NFE.set_index('HUGO_SYMBOL')['High Score'].to_dict()
    NFE_True['High Score'] = NFE_True[item].map(Max_dict)
    
    count_dict = LUAD_NFE.set_index('HUGO_SYMBOL')['Count'].to_dict()
    NFE_True['Count'] = NFE_True[item].map(count_dict)
    
    median_dict = LUAD_NFE.set_index('HUGO_SYMBOL')['Median'].to_dict()
    NFE_True['Median'] = NFE_True[item].map(median_dict)
    
    NFE_True['Metabolic'] = (NFE_True[item]).isin(KEGG_M['GENE'])
    
    NFE_True['Metabolic'] = (NFE_True[item]).isin(KEGG_M['GENE'])
    
    print(NFE_True, '\n')
    
```
