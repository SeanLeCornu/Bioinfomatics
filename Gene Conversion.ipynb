```python
import pandas as pd 
```

```python
name_id = pd.read_table('uniprot-filtered-proteome_AUP000005640_reviewed_9606.tab', sep='\t')

name_id
```

```python
name_id = name_id.fillna('-')

name_id
```

```python
STRINGid = pd.read_csv('TCGA Barcodes/CEMI Analysis/STRING Files/9606.protein.links.v11.0.txt', sep=" ", header=None)
STRINGid.columns = ["Protein 1", "Protein 2", "Score"]


STRINGid
```

```python
convert_dict = name_id.set_index('Cross-reference_STRING')['Gene_names_primary'].to_dict()

convert_dict = {key.strip(';'): item.strip() for key, item in convert_dict.items()}

convert_dict
```

```python
conv_STRING = pd.Series(STRINGid['Protein 1'].map(convert_dict))

conv_STRING
```

```python
conv2_STRING = pd.Series(STRINGid['Protein 2'].map(convert_dict))

conv2_STRING
```

```python
final = pd.concat([conv_STRING, conv2_STRING], axis = 1)

final
```

```python
final = final.drop([0,1])

final

# just rerun to remove error 
```

```python
final.to_csv('STRING Interactions.csv', index=False)
```
