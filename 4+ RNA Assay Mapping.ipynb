```python
import pandas as pd 
```

```python
LUAD_NFE = pd.read_csv('LUAD NFE + Median.csv', sep=",")

LUAD_NFE
```

```python
LUAD_4 = LUAD_NFE.loc[LUAD_NFE['High Score'] > 4]

LUAD_4
```

```python
LUAD_Over_4 = LUAD_4['HUGO_SYMBOL']

LUAD_Over_4
```

```python
LUAD_Assay = pd.read_csv('TCGA Barcodes/LUAD Assays Normalised.csv', sep=",")

LUAD_Assay
```

```python
LUAD_Assay['Filter'] = LUAD_Assay.gene_id.isin(LUAD_4.HUGO_SYMBOL).astype(bool)

LUAD_Assay
```

```python
Filt_LUAD_Assay = LUAD_Assay.loc[LUAD_Assay['Filter'] == True]

Filt_LUAD_Assay
```

```python
Filt_LUAD_Assay.to_csv('Filtered_LUAD_Assay_4+.csv')
```
